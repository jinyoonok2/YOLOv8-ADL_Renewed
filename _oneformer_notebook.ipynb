```python
import os
import yaml

DATASET_DIR = "./datasets/PlantVillage_apple_mask"

YAML_PATH = os.path.join(DATASET_DIR, "APPLE_DATA_YAML.yaml")

APPLE_TRAIN_PATH = os.path.join(DATASET_DIR, "APPLE-TRAIN-YOLO")
TRAIN_PATH_IMAGE = os.path.join(APPLE_TRAIN_PATH, "images")
TRAIN_PATH_MASK = os.path.join(APPLE_TRAIN_PATH, "masks")
TRAIN_PATH_LABEL = os.path.join(APPLE_TRAIN_PATH, "labels")

APPLE_TEST_PATH = os.path.join(DATASET_DIR, "APPLE-TEST-YOLO")
TEST_PATH_IMAGE = os.path.join(APPLE_TEST_PATH, "images")
TEST_PATH_MASK = os.path.join(APPLE_TEST_PATH, "masks")
TEST_PATH_LABEL = os.path.join(APPLE_TEST_PATH, "labels")

ACTIVE_LEARNING_PATH = os.path.join(DATASET_DIR, "APPLE_ACTIVE_LEARNING")

# Load class names from a YAML file
with open(YAML_PATH, 'r') as file:
    class_names = yaml.safe_load(file)
# Assuming the class names are under the 'names' key in your YAML file
CLASSES = class_names['names']
# Create id2label dictionary
ID2LABEL = {idx: label for idx, label in enumerate(CLASSES)}
# Increment each key by 1
id2label = {class_id + 1: class_name for class_id, class_name in ID2LABEL.items()}
```

# Dataload

```python
from transformers import AutoProcessor, AutoModelForUniversalSegmentation
from torch.optim import AdamW
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import os
import numpy as np
import torch
import matplotlib.pyplot as plt
import albumentations as A
import cv2

ADE_MEAN = np.array([123.675, 116.280, 103.530]) / 255
ADE_STD = np.array([58.395, 57.120, 57.375]) / 255
```

```python
class CustomDataset(Dataset):
    def __init__(self, image_dir, mask_dir, processor, transform=None):
        self.processor = processor
        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.images = sorted([img for img in os.listdir(image_dir) if img.endswith('.jpg')])
        self.masks = sorted([mask for mask in os.listdir(mask_dir) if mask.endswith('_mask.png')])
        self.transform = transform
    def __getitem__(self, idx):
        # Load image and mask

        # Load image and mask
        image_path = os.path.join(self.image_dir, self.images[idx])
        mask_path = os.path.join(self.mask_dir, self.masks[idx])

        image = Image.open(image_path).convert("RGB")  # Ensure image is in RGB format
        mask = Image.open(mask_path)
        mask = np.array(mask)  # Convert mask to numpy array

        # Apply transformations to the image and mask
        if self.transform:
            augmented = self.transform(image=np.array(image), mask=mask)  # Convert PIL image to numpy array
            image = augmented['image']
            mask = augmented['mask']

        # Use processor to convert this to a list of binary masks, labels, text inputs, and task inputs
        inputs = self.processor(images=image, segmentation_maps=mask, task_inputs=["semantic"], return_tensors="pt")
        inputs = {k: v.squeeze() if isinstance(v, torch.Tensor) else v[0] for k, v in inputs.items()}

        return inputs

    def __len__(self):
        return len(self.images)

    def get_class_id_for_image(self, idx):
        """
        Get the class ID for the mask of a specific image.
        Assumes that each mask contains only one class apart from the background.
        Args:
            idx (int): Index of the image in the dataset.

        Returns:
            int: Class ID of the mask.
        """
        mask_path = os.path.join(self.mask_dir, self.masks[idx])
        mask = Image.open(mask_path).convert('L')
        mask_array = np.array(mask)

        # Use np.max to get the highest value in the mask, assuming it represents the class ID
        class_id = np.max(mask_array)
        return class_id

def create_train_transform():
    train_transform = A.Compose([
        A.HorizontalFlip(p=0.5),
        A.Normalize(mean=ADE_MEAN, std=ADE_STD),
        A.Rotate(limit=35, p=0.5),  # Random rotation between -35 and +35 degrees, 50% chance to apply rotation
    ])
    return train_transform
```

```python
processor = AutoProcessor.from_pretrained("shi-labs/oneformer_coco_swin_large")
model = AutoModelForUniversalSegmentation.from_pretrained("shi-labs/oneformer_coco_swin_large", is_training=True)
processor.image_processor.num_text = model.config.num_queries - model.config.text_encoder_n_ctx
processor.image_processor.do_resize = False
```

# Train the Model

```python
import torch
from transformers import AutoModelForUniversalSegmentation, AdamW
import os
from pathlib import Path
from tqdm import tqdm
```

```python
def setup_model_and_optimizer(pretrained_model_name="shi-labs/oneformer_coco_swin_large", learning_rate=5e-5):
    model = AutoModelForUniversalSegmentation.from_pretrained(pretrained_model_name, is_training=True)
    optimizer = AdamW(model.parameters(), lr=learning_rate)
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.to(device)
    return model, optimizer, device

def train_model(model, optimizer, device, train_dataloader, checkpoint_dir, num_epochs=12):
    model.train()
    
    for epoch in range(num_epochs):  # Start from 0 to num_epochs
        total_loss = 0
        num_batches = 0
        progress_bar = tqdm(enumerate(train_dataloader), total=len(train_dataloader))
        for i, batch in progress_bar:
            optimizer.zero_grad()
            batch = {k: v.to(device) for k, v in batch.items()}
            outputs = model(**batch)
            loss = outputs.loss
            total_loss += loss.item()
            num_batches += 1
            loss.backward()
            optimizer.step()
            progress_bar.set_description(f"Epoch {epoch}/{num_epochs - 1} Loss: {loss.item()}")
        
        avg_loss = total_loss / num_batches
        print(f"Epoch: {epoch}, Avg Loss: {avg_loss}")
    
    # Save final model
    final_model_path = os.path.join(checkpoint_dir, 'final_model.pt')
    torch.save(model.state_dict(), final_model_path)
    print(f"Final model saved: {final_model_path}")
    
    return final_model_path  # Return the path of the saved final model
```

# Model Inference

```python
import numpy as np
from PIL import Image
from transformers import AutoModelForUniversalSegmentation
from torchvision.transforms.functional import to_pil_image
import random
from shutil import copyfile
```

```python
# Global variable to store the initial lengths of selected images for each class
initial_lengths = {}

def load_model(checkpoint_path):
    model = AutoModelForUniversalSegmentation.from_pretrained("shi-labs/oneformer_coco_swin_large", is_training=True)
    checkpoint = torch.load(checkpoint_path)
    # Directly load the state dictionary without accessing 'model_state_dict'
    model.load_state_dict(checkpoint)
    model.model.is_training = False
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()
    return model, device

def process_and_save_mask(image_path, mask_output_path, model, processor, device):
    image = Image.open(image_path).convert('RGB')
    inputs = processor(images=image, task_inputs=["semantic"], return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = model(**inputs)
    semantic_segmentation = processor.post_process_semantic_segmentation(outputs, target_sizes=[(256, 256)])[0]
    
    # Check if there are any detections other than background
    unique_values = np.unique(semantic_segmentation.cpu().numpy())
    if len(unique_values) == 1 and unique_values[0] == 0:
        print(f"No detections except for background in {os.path.basename(image_path)}. Mask creation skipped.")
        return  # Skip saving the mask

    # Convert tensor to PIL image for saving
    segmentation_mask_image = to_pil_image(semantic_segmentation.byte())
    segmentation_mask_image.save(mask_output_path)
    print(f"Inferenced: {os.path.basename(image_path)} -> Created mask at: {mask_output_path}")


def infer(image_dir, output_dir, model, processor, device, inference_round=1):
    global initial_lengths
    for class_name in CLASSES:
        class_dir = os.path.join(image_dir, class_name)
        output_class_dir = os.path.join(output_dir, class_name)
        os.makedirs(output_class_dir, exist_ok=True)
        
        if not os.path.isdir(class_dir):
            print(f"Class directory {class_dir} does not exist.")
            continue

        images = os.listdir(class_dir)
        if inference_round == 1:
            initial_lengths[class_name] = max(1, int(len(images) * 0.2))
        
        selected_images = random.sample(images, min(len(images), initial_lengths[class_name]))

        if not selected_images:
            print(f"No more images available for inference in {class_name} class.")
            continue

        for img_name in selected_images:
            img_path = os.path.join(class_dir, img_name)
            mask_output_path = os.path.join(output_class_dir, os.path.splitext(img_name)[0] + '_mask.png')
            process_and_save_mask(img_path, mask_output_path, model, processor, device)
```

# Result Correction and Distribution into Dataset

```python
import numpy as np
from PIL import Image
import os
import shutil

def correct_mask_pixels(output_path, id2label):
    # Iterate over each class name and its corresponding ID in id2label
    for class_id, class_name in id2label.items():
        class_folder_path = os.path.join(output_path, class_name)
        if not os.path.isdir(class_folder_path):
            print(f"Class folder {class_folder_path} does not exist.")
            continue
        
        # List all mask files in the current class folder
        mask_files = [f for f in os.listdir(class_folder_path) if f.endswith('.png')]
        total_masks = len(mask_files)
        corrected_masks = 0  # Counter for corrected masks
        
        for mask_file in mask_files:
            mask_path = os.path.join(class_folder_path, mask_file)
            mask = np.array(Image.open(mask_path))
            
            # Check if correction is needed
            needs_correction = np.any((mask != 0) & (mask != class_id))
            
            # Correct the mask pixels
            corrected_mask = np.where((mask != 0) & (mask != class_id), class_id, mask)
            
            # Save the corrected mask back to disk only if correction was needed
            if needs_correction:
                corrected_mask_image = Image.fromarray(corrected_mask.astype(np.uint8))
                corrected_mask_image.save(mask_path)
                corrected_masks += 1  # Increment the corrected masks counter
            
        # Calculate the percentage of corrected masks
        correction_percentage = (corrected_masks / total_masks) * 100 if total_masks > 0 else 0
        
        # Print the summary for this class
        print(f"Class '{class_name}': Total Masks = {total_masks}, Corrected Masks = {corrected_masks}, Correction Percentage = {correction_percentage:.2f}%")

def move_files_to_dataset(active_learning_path, train_path, output_path):
    # Define subdirectories for images and masks within the train path
    train_images_path = os.path.join(train_path, 'images')
    train_masks_path = os.path.join(train_path, 'masks')
    
    # Ensure the subdirectories exist
    os.makedirs(train_images_path, exist_ok=True)
    os.makedirs(train_masks_path, exist_ok=True)

    # Iterate over each class in the active learning path
    for class_name in os.listdir(active_learning_path):
        class_folder_path = os.path.join(active_learning_path, class_name)
        output_class_folder_path = os.path.join(output_path, class_name)
        
        if not os.path.isdir(class_folder_path) or not os.path.isdir(output_class_folder_path):
            print(f"Skipping {class_name} due to missing directory.")
            continue
        
        # List all mask files in the output class folder
        mask_files = [f for f in os.listdir(output_class_folder_path) if f.endswith('_mask.png')]
        
        for mask_file in mask_files:
            # Construct the corresponding image file name
            image_file_name = mask_file.replace('_mask.png', '.jpg')
            image_file_path = os.path.join(class_folder_path, image_file_name)
            mask_file_path = os.path.join(output_class_folder_path, mask_file)
            
            # Check if the corresponding image file exists
            if os.path.isfile(image_file_path):
                # Move the mask file to the train masks directory
                shutil.move(mask_file_path, os.path.join(train_masks_path, mask_file))
                # Move the image file to the train images directory
                shutil.move(image_file_path, os.path.join(train_images_path, image_file_name))
            else:
                print(f"Image file {image_file_name} not found for mask {mask_file}. Skipping.")
```

```python
active_path = './datasets/PlantVillage_apple_mask/APPLE_ACTIVE_LEARNING'  # Update with your actual path
output_path = './datasets/PlantVillage_apple_mask/OUTPUT_PATH'  # Update with your actual path
base_checkpoint_dir = 'one_former/train'

# Create the output directory if it doesn't exist
os.makedirs(output_path, exist_ok=True)
previous_num_files = None  # Initialize variable to keep track of the number of files in the previous iteration

# Repeat the process for 6 iterations
for iteration in range(1, 11):
    print(f"Starting iteration {iteration}...")
    
    # Check the current number of files in the active learning path
    current_num_files = sum([len(files) for r, d, files in os.walk(active_path)])
    
    # Exit the loop if there are no files or if the number of files hasn't changed
    if current_num_files == 0 or current_num_files == previous_num_files:
        print("No more files to process or no change in file count. Exiting loop.")
        break
    
    # Update previous_num_files for the next iteration
    previous_num_files = current_num_files
    
    # Dynamically set the checkpoint_dir for the current iteration
    checkpoint_dir = f"{base_checkpoint_dir}-final-{iteration}"
    os.makedirs(checkpoint_dir, exist_ok=True)  # Ensure the directory exists
    
    # Reload the dataset and dataloader for the current iteration
    train_dataset = CustomDataset(TRAIN_PATH_IMAGE, TRAIN_PATH_MASK, processor)
    train_dataloader = DataLoader(train_dataset, batch_size=2, shuffle=True)
    
    # Setup model and optimizer for the current iteration
    model, optimizer, device = setup_model_and_optimizer()
    
    # Train the model
    final_model_path = train_model(model, optimizer, device, train_dataloader, checkpoint_dir)
    
    # Load the model for inference using the final model from the current iteration
    model, device = load_model(final_model_path)
    
    # Perform inference
    infer(active_path, output_path, model, processor, device, inference_round=iteration)
    
    # Correct misclassified mask pixels
    correct_mask_pixels(output_path, id2label)
    
    # Move the processed files to the dataset for the next round
    move_files_to_dataset(active_path, APPLE_TRAIN_PATH, output_path)
    
    print(f"Iteration {iteration} completed.")
# If the loop exits due to break, you can handle post-loop actions here (if any)
print("Process completed.")
```
